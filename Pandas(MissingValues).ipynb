# Handling missing Data

Make given string read as NaN values

```python
df=pd.read_csv("test.csv", na_values="male")
df
```

make multiple values read as NaN values

```python
df=pd.read_csv("test.csv", na_values=["male", "female"])
df
```

make column values NaN.(by using dictionary)

```python
df=pd.read_csv("test.csv", na_values={"Age":34.5, "Sex":"male"})
df
```

keep "null","no values" same.

```python
df=pd.read_csv("test.csv",keep_default_na=False)
df
```

Check the NaN values........
#in place of NaN there will come True

```python
df.isnull()
```

Find the total null values.

```python
df.isnull().sum()
```

notnull

```python
df.notnull().sum()
```

Drop the NaN values-------> by rows

```python
df=pd.read_csv("test.csv")
df.dropna()
```

```python
df=pd.read_csv("test.csv")
df.dropna(axis=1)
```
